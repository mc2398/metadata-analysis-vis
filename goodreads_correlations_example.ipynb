Program to produce Pearson's R correlation coefficients and p-values for data

```python
import pandas as pd
```

```python
from scipy.stats import pearsonr
```

This script works for analyzing data about the presence/absence of any two factors (encoded as binary values of 1s and 0s) imported as a csv.  For calculating correlations between more than 2 factors at a time, you would need to use regression instead.

```python
#specify the csv file path- update as needed
csv_file_path = '/content/correlation_comparisons_lgbt_x_queer.csv'
```

```python
# script to read the csv into memory and convert it to a pandas data frame
lgbt_queer_df = pd.read_csv(csv_file_path)
```

```python
#defining a function to produce pearson's r and p values for a correlation of 2 values
# using the scipystats library and pandas
def p_value_calc(df, col1, col2):
  corr, p_value = pearsonr(df[col1], df[col2])
  return corr, p_value
```

```python
# passing the function the data frame for the csv I want to analyze, and the
# specifc column names to correlate- must match exactly what is in the csv
correlation, p_val = p_value_calc(lgbt_queer_df, 'lgbt_subject', 'queer_subject')
print(f'{correlation}')
print(f'{p_val}')
```
